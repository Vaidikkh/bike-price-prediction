```python
tvs_first_owner_under_34_df = tvs_first_owner_df[tvs_first_owner_df['price']<=50000]
tvs_first_owner_under_34_df
```

```python
royal_enfield = df[df['brand']=='Royal Enfield']
royal_enfield
```

```python
royal_enfield_under_20000 = royal_enfield[royal_enfield['price']<=200000]
royal_enfield_under_20000
```

```python
royal_enfield_under_20000_first_owner = royal_enfield_under_20000[royal_enfield_under_20000['owner']=='First Owner']
royal_enfield_under_20000_first_owner
```

```python
royal_enfield_under_20000_first_owner_under_50k = royal_enfield_under_20000_first_owner[royal_enfield_under_20000_first_owner['kms_driven']<=50000.0]
royal_enfield_under_20000_first_owner_under_50k
```

```python
royal_enfield_under_20000_first_owner_under_50k_age2= royal_enfield_under_20000_first_owner_under_50k[royal_enfield_under_20000_first_owner_under_50k['age']==2]
royal_enfield_under_20000_first_owner_under_50k_age2 
```
